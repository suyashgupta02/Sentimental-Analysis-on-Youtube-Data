```python
import pandas as pd
import numpy as np
```

```python
UScomments = pd.read_csv('UScomments.csv', error_bad_lines=False)
```

```python
UScomments
```

```python
def remove_special(text):
    import string
    return text.translate(str.maketrans('','',string.punctuation))
```

```python
UScomments = UScomments.fillna({'comment_text':''})
```

```python
UScomments['comments_clean'] = UScomments['comment_text'].apply(remove_special)
```

```python
UScomments['analysis'] = ""
```

```python
UScomments
```

```python
from textblob import TextBlob

for i in range (0,245212):
    temp = TextBlob(UScomments['comments_clean'].tolist()[i]) 
    if temp.sentiment.polarity > 0:
        UScomments['analysis'][i]= 'positive'
    elif temp.sentiment.polarity == 0:
        UScomments['analysis'][i] ='neutral'
    else:
        UScomments['analysis'][i]= 'negative'
```

```python
UScomments.head()
```

```python
UScomments['rating'] = ''
UScomments
```

```python
x=0

for j in range (1,245212):
    
    if UScomments['video_id'][j]==UScomments['video_id'][j-1]:
        if UScomments['analysis'][j]=='positive':
            x = x+1
            UScomments['rating'][j]=x
        if UScomments['analysis'][j]=='negative':
            x = x-1
            UScomments['rating'][j]=x
        if UScomments['analysis'][j]=='neutral':
            UScomments['rating'][j]=x
    else:
        x=0
        
        
    
```

```python
UScomments
```

```python
NewDataframe = pd.DataFrame(index=range(0,2800),columns=['video_id','rating'])
```

```python
z=0
for k in range (0,245211):
    
    if UScomments['video_id'][k]!=UScomments['video_id'][k+1]:
        NewDataframe['video_id'][z] = UScomments['video_id'][k]
        NewDataframe['rating'][z] = UScomments['rating'][k]
        z=z+1
    else:
        pass
```
